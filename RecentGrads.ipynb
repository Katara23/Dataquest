This dataset contains job outcomes for recent college graduates.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv")
recent_grads = recent_grads.dropna()
print(recent_grads.shape[0],"rows")
```

The graph below is a scatter matrix that shows the relationship between the unemployment rate and the number of women as a share of the total number of people in the dataset.

```python
sns.set_style("white")
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.hist(recent_grads["ShareWomen"])
ax2.scatter(recent_grads["Unemployment_rate"],recent_grads["ShareWomen"])
ax3.scatter(recent_grads["ShareWomen"],recent_grads["Unemployment_rate"])
ax4.hist(recent_grads["Unemployment_rate"])

plt.subplots_adjust(wspace=0, hspace=0)
ax1.xaxis.set_visible(False)
ax2.xaxis.set_visible(False)
ax2.yaxis.set_visible(False)
ax4.yaxis.set_visible(False)

ax1.set_yticklabels([0, 5, 10, 15, 20, 25, 30])
ax3.set_yticklabels([0.00, 0.05, 0.10, 0.15])
ax3.set_xticklabels([0.0, 0.2, 0.4, 0.6, 0.8], rotation=90)
ax4.set_xticklabels([0.00, 0.05, 0.10, 0.15, 0.20], rotation=90)

ax1.set_ylim(0,30)
ax2.set_ylim(0.0,1.0)
ax3.set_xlim(0.0,1.0)
ax3.set_ylim(0.0,0.20)
ax4.set_xlim(0.0,0.20)

ax1.set(ylabel="Share Women")
ax3.set(xlabel="Share Women",ylabel="Unemployment Rate")
ax4.set(xlabel="Unemployment Rate")

sns.plt.show()
```

The graph below shows the relationshp between the number of males and females for each major within the Arts major category. Women outnumber men in every category except Miscellaneous Fine Arts and Film Video And Photographic Arts.

```python
sns.set_style("whitegrid")
recent_grads["ShareMen"] = recent_grads["Men"] / recent_grads["Total"]
arts = recent_grads[recent_grads["Major_category"] == "Arts"]
major_names = arts["Major"].tolist()
locs = np.arange(len(major_names))
offset_locs = [x+0.35 for x in locs]

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
bar1 = ax.bar(locs, arts["ShareMen"].tolist(),0.35)
bar2 = ax.bar(offset_locs, arts["ShareWomen"],0.35,color="green")
ax.set_xticklabels(major_names,rotation=90)
ax.set_xticks(offset_locs)
plt.legend((bar1, bar2), ("Share Men", "Share Women"), loc="upper left")
sns.plt.show()
```
